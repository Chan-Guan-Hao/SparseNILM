```python
import pandas as pd
import matplotlib
```

```python
data = pd.read_csv('datasets\synthetic_REDD_test_I.csv', parse_dates = True, index_col = 0)
test_data = data / 1000
test_data.to_csv('datasets\synthetic_REDD_test_II.csv')
```

```python
test_data
```

```python
disaggregated = pd.read_csv('datasets\disaggregated.csv', index_col = 0)
```

```python
disaggregated.columns = test_data.columns[:-1]
```

```python
test_data[:1000].reset_index(drop = True)
```

```python
disaggregated.columns = ['disaggregated_' + col for col in disaggregated.columns]
test_data.columns = ['ground_truth_' + col for col in test_data.columns]
```

```python
print([x[14:] for x in disaggregated.columns])
```

```python
for i in disaggregated.columns:
    result = pd.concat([test_data[:10000].reset_index(drop = True)['ground_truth_' + i[14:]], disaggregated[:10000][i]], axis = 1)
    result.plot(figsize = (15, 10))
```
